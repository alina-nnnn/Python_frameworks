#Manipulating tensors in Numpy

Tensor slicing

You may select between any two indices along each tensor axis

```
my_slice = x_train[:, :14, :14] # the top-left corner of all images
digit_slice = my_slice[17]
plt.imshow(digit_slice, cmap=plt.cm.binary)
plt.show()
```

```
my_slice.shape
```

```
my_slice = x_train[:, :14, 14:] # the top-right corner of all images
digit_slice = my_slice[17]
plt.imshow(digit_slice, cmap=plt.cm.binary)
plt.show()
```

```
my_slice = x_train[:, 14:, :14] #  the bottom-left corner of all images
digit_slice = my_slice[17]
plt.imshow(digit_slice, cmap=plt.cm.binary)
plt.show()
```

```
my_slice_up = x_train[:, :16, :] # the top digit
digit_slice = my_slice_up[17]
plt.imshow(digit_slice, cmap=plt.cm.binary)
plt.show()
```

```
my_slice_centre = x_train[:, 7:-7, 7:-7] # the centre
centre_slice = my_slice_centre[17]
plt.imshow(centre_slice, cmap=plt.cm.binary)
plt.show()
```
